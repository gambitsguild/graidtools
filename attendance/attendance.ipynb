```python
import re
import pandas as pd
from datetime import datetime

data = pd.read_csv('gambits_attendance_21-10-03.csv')

N_RAIDS = 8

data.index = data['Name']
data = data.drop(columns=["Name", "Att.%"]).fillna(0)
# Drop Karazan attendance
data = data.drop(columns=list(data.filter(regex="Kar")))
# Extract the dates from the names
data = data.rename(columns=lambda x: datetime.strptime(re.search("\d+/\d+", x)[0], "%d/%m").replace(year=2021).date())
# Group raids on same day and take the mean
data = data.groupby(lambda x:x, axis=1).mean()
# Sort dates descending
data = data.reindex(sorted(data.columns, reverse=True), axis=1)

# Take latest N_RAIDS raids
data = data.iloc[:, :N_RAIDS]
data["Attendance"] = data.iloc[:, :].sum(axis=1) / N_RAIDS

data.to_csv('parsed_attendance.csv')

data
```
